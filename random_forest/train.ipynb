```python
import sys
import os
sys.path.append('../')
from random_forest import *
from utils.data_processing import load_data
from joblib import dump

PEROID_TYPE = '15minutely'

if PEROID_TYPE == 'daily':
    start_date_train = "2000-01-30"
    end_date_train = "2023-01-01"

    start_date_test = "2023-01-01"
    end_date_test = "2024-01-01"
elif PEROID_TYPE == '15minutely':
    start_date_train = "2000-01-01"
    end_date_train = "2023-05-01"

    start_date_test = "2023-05-01"
    end_date_test = "2024-01-01"
elif PEROID_TYPE == 'minutely':
    start_date_train = "2022-01-30"
    end_date_train = "2023-08-06"

    start_date_test = "2023-08-06"
    end_date_test = "2024-01-01"

tics = [
    "AAPL",
    "GOOG",
    "AMZN",
    "NVDA",
    "TSLA",
    "AMD",
    "INTC",
    "ATVI",
    "PYPL",
    "TTD",
    "EA",
    ]
x_train, y_train, x_test, y_test, models = {}, {}, {}, {}, {}

input_items = [
    
    'RSI5',
    'RSI10',
    'RSI20',
    'RSI60',
    
    'PD_SMA5_Close',
    'PD_SMA10_Close',
    'PD_SMA20_Close',
    'PD_SMA60_Close',
    
    'PD_SMA5_SMA10', 
    'PD_SMA5_SMA20', 
    'PD_SMA5_SMA60',
    
    'PD_SMA10_SMA20',
    'PD_SMA10_SMA60',
    
    'PD_SMA20_SMA60',  
    
    'PD_RSI5_RSI10',
    'PD_RSI5_RSI20',
    'PD_RSI5_RSI60',
    
    'PD_RSI10_RSI20',
    'PD_RSI10_RSI60',
    
    'PD_RSI20_RSI60',
    
    'SL3CLOSE',
    'SL5CLOSE',
    'SL10CLOSE',
    'SL20CLOSE',
    'SL60CLOSE',
    
    'SL3SMA5',
    'SL5SMA5',
    'SL10SMA5',
    'SL20SMA5',
    'SL60SMA5',
    
    'SL3SMA10',
    'SL5SMA10',
    'SL10SMA10',
    'SL20SMA10',
    'SL60SMA10',
    
    'SL3SMA20',
    'SL5SMA20',
    'SL10SMA20',
    'SL20SMA20',
    'SL60SMA20',
    
    'SL3SMA60',
    'SL5SMA60',
    'SL10SMA60',
    'SL20SMA60',
    'SL60SMA60',
    
    'SL3RSI5',
    'SL5RSI5',
    'SL10RSI5',
    'SL20RSI5',
    'SL60RSI5',
    
    'SL3RSI10',
    'SL5RSI10',
    'SL10RSI10',
    'SL20RSI10',
    'SL60RSI10',
    
    'SL3RSI20',
    'SL5RSI20',
    'SL10RSI20',
    'SL20RSI20',
    'SL60RSI20',
]

target_items = [
    'FR1TF',
    'FR3TF',
    'FR5TF',
    'FR10TF',
]

for tic in tics:
    print('-'*50 + tic +'-'*50)
    
    path = "../data/" + PEROID_TYPE + "/" + tic + ".csv"
    
    train_data = load_data(path, start_date_train, end_date_train)
    train_data = train_data.dropna()
    test_data = load_data(path, start_date_test, end_date_test)
    test_data = test_data.dropna()
    
    x_train[tic] = train_data[input_items]
    x_test[tic] = test_data[input_items]
    
    for target_item in target_items:
        print('*'*30 + target_item +'*'*30)
        y_train[tic] = train_data[target_item]
        y_test[tic] = test_data[target_item]
        print('len(y_train[tic])',len(y_train[tic]),'len(y_test[tic])',len(y_test[tic]))
        
        models[tic] = train_random_forest(x_train[tic], y_train[tic])
        
        random_accuracy = evaluate(models[tic], x_train[tic], y_train[tic])
        print('Train Accuracy = {:0.2f}%.'.format(random_accuracy))
        random_accuracy = evaluate(models[tic], x_test[tic], y_test[tic])
        print('Test Accuracy = {:0.2f}%.'.format(random_accuracy))
        
        if not os.path.exists('model/' + tic):
            os.makedirs('model/' + tic)
            
        if not os.path.exists('model/' + tic + '/' + PEROID_TYPE):
            os.makedirs('model/' + tic + '/' + PEROID_TYPE)
        
        dump(models[tic], 'model/' + tic + '/' + PEROID_TYPE + '/' + target_item + '.joblib')
    
    
    
```
